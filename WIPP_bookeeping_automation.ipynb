#### Retrieve assets

```python
url_asset = "https://api.opensea.io/api/v1/assets"

querystring_asset = {"order_direction":"desc","offset":"0","limit":"20",\
                     'owner':WIP_OpenSea_address}
#                     'asset_contract_address':WIP_asset}

header_asset = {'X-API-KEY': WIP_API_key}

response_asset = requests.request("GET", url_asset, headers=header_asset, params=querystring_asset)

response_asset.json()
```

#### Retrieve events

```python
url_event = "https://api.opensea.io/api/v1/events"

querystring_event = {"only_opensea":"false","offset":"0","limit":"50",\
                     'account_address':WIP_OpenSea_address}

header_event = {"Accept": "application/json"}

response_event = requests.request("GET", url_event, headers=header_event, params=querystring_event)

response_event.json()
```

#### Retrieve orders (from OpenSea's *orderbook*)

```python
url_order = "https://api.opensea.io/wyvern/v1/orders"

querystring_order = {"bundled":"false","include_bundled":"false","include_invalid":"false","limit":"20","offset":"0","order_by":"created_date","order_direction":"desc"}

header_order = {"Accept": "application/json"}

response_order = requests.request("GET", url_order, headers=header_order, params=querystring_order)

response_order.json()
```
